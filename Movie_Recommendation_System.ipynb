```python
# creating Helping function
import ast
def convert(obj): ## '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"},
#  {"id": 14, "name": "Fantasy"},
#  {"id": 878, "name": "Science Fiction"}]'--> This will go inside this function
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
    
```

```python
# In cast we need top three actor not all

def convert3(obj):
    l=[]
    cnt=0
    for i in ast.literal_eval(obj):
        if cnt !=3:
           l.append(i['name'])
           cnt+=1
        else:
           break
    return l
```

```python
# Crew
# i want the dictionary whose job is director 
# i do not care about others job
def fetch_director(obj):
     l=[]
     for i in ast.literal_eval(obj):
         if i['job']=='Director':
             l.append(i['name'])
             break
     return l

# Description of code
# First we enter in the dictornary
# We want the name whose job is derector
# As in every movie only one director so we don't have to explore futher dictonary
# For that we use break 
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

# ['loves','loving','loved']--> this becoming 3 different attribute which is an issue
# I want evrything in one formate
```

```python
# Helping function
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

Vectorization

```python
# Frist thing calculating the similarity score between two movies
# In general calculating how much simliarity of 1 movies with every moves
# As it is textual data we can not apply some calculation to calculate the similarity between 2 movies
# We convert the text into vector(Basically codinates in N-Dimensiona)
# This is called text vectorization
# Bag of tags
# Converting evry movie in vector
# Before that we combine all the tags of movies and
# Find the most 5000 frequently occure words in that combined tag
# According to that words we take first movie and see how many time 5k words occure in movie 1 , that is first vector
# same goes for all the movies, this is vector is calculated

```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray()

# Converting into numpy array beacuse is give spase matrix,
# in which mostly 0 are present
```

```python
cv.get_feature_names_out()
```

```python
words=cv.get_feature_names_out()
print(words[:100])
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0] # First mmovie in vector form

# Corpus--> sare tags ko a ek me add krke jo string bana ushko bolte hsi
```

```python
# Now we have converted all the movies into vector
# We have 4806 movies and vector also 4806
# And each vector is having 5000 numbers
# Now we have to calculate the distance of every movie with every other movie
# Distance jitna jada hoga similarity utna km hoga movie ke bich
# We will not calculate the euclindine distance(calculating the distance between two cordinate)
# Ex: (6,6),(2,4)-->POint to point distance(Not preferable for N-D)
# Instead of this we will caluclate the cosine distance(for High Dimenetional data)
# Distance between two movies, means the angle between two movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# PAssing the 4806 vector into this

cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
# Here movie ka hr movie ke sth distance niklana hai
# 1 movie ka 4806 movies ke sth distance niklana hai
```

```python
similarity=cosine_similarity(vectors)
```

```python
# Now we are going to make one function
# Now the thing is we will given one movie from this 4806 
# And we have to give 5 similar movies realted to given movie

# Funda:
# First we take movie exam-batman
# the vector we have  already find,
# Basically that vector of batsman contain all distances from 4806 movies
# We will sort that vector or distance
# Then we will take the 5 movies/distances which have the more simaliarity with batsman movie


# Step-01
# Finding the index of given movie in 4806





def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    # Now one by one we will print the index of the movies
    for i in movies_list:
       # print(new_df.iloc[i[0]].title)
        print(i[0])


   
```
